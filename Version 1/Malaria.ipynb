```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.datasets import mnist
import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from PIL import Image
import os
```

```python
from PIL import Image
import os
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Assuming you have df, image_size, and label_encoder defined

images = []
image_size = (100, 75)
for index, row in df.iterrows():
    image_path = row['FullPath']
    
    # Check if the file is an image file based on its extension
    valid_image_extensions = ('.png', '.jpg', '.jpeg', '.gif', '.bmp')
    if os.path.splitext(image_path)[-1].lower() in valid_image_extensions:
        image = Image.open(image_path)
        image = image.resize(image_size)  # Resize images if needed
        image_array = np.array(image, dtype=np.float32) / 255.0  # Normalize pixel values to [0, 1]
        images.append(image_array)

# Convert the list of image arrays to a NumPy array
X = np.array(images)

# Step 3: Extract labels and encode them
y = df['target'].values
y = y[:-2]
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_one_hot = to_categorical(y_encoded)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.2, random_state=42,stratify=y)
```

```python
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], X_train.shape[2], 3))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], X_test.shape[2], 3))
```

```python
model = Sequential()

# Add layers to the model
model.add(Conv2D(32, kernel_size=(3, 3), input_shape=(75, 100, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Flatten())

model.add(Dropout(0.25))

# Adjust the number of units in the output layer based on your classification task
model.add(Dense(2, activation='softmax'))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Define callbacks
es = EarlyStopping(monitor='val_accuracy', min_delta=0.01, patience=2, verbose=1, restore_best_weights=True)
mc = ModelCheckpoint("./bestmodel.h5", monitor="val_accuracy", verbose=1, save_best_only=True)

# Train the model with callbacks
history = model.fit(X_train, y_train, epochs=50, validation_split=0.3)
```
